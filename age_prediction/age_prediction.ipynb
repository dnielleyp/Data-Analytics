I will be implementing a CNN model to perform age prediction on a set of facial images.

The general CNN architecture consists of:
1. input layer
2. convolutional layer
3. pooling layer
4. fully connected layer 
5. output layer

The outputs from each layer serves as an input to the following layer up until the final output layer.

Despite age being a numerical data, I will be performing a categorical prediction to predict the age range of the person in the image. This is because someone's facial features are not solely reliant on their age since external factors play a huge role on a person's appearance.
